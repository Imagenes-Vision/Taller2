```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
```

## Visualizaci√≥n

```python
rondelle= cv2.imread('img_lab2/RONDELLE.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(rondelle,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([rondelle], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
angio= cv2.imread('img_lab2/ANGIO.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(angio,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
hist = cv2.calcHist([angio], [0], None, [256], [0, 256])
plt.plot(hist)
```

## Umbralizaci√≥n Simple

```python
thres_value, thres_image_0 = cv2.threshold(rondelle, 0, 255, cv2.THRESH_BINARY)
plt.imshow(thres_image_0,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

hist = cv2.calcHist([thres_image_0], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
thres_value, thres_image_128 = cv2.threshold(rondelle, 128, 255, cv2.THRESH_BINARY)
plt.imshow(thres_image_128,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

hist = cv2.calcHist([thres_image_128], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
thres_value, thres_image_255 = cv2.threshold(rondelle, 255, 255, cv2.THRESH_BINARY)
plt.imshow(thres_image_255,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

hist = cv2.calcHist([thres_image_255], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
thres_image = cv2.inRange(rondelle, 70, 140)
plt.imshow(thres_image,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

hist = cv2.calcHist([thres_image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
