# Ecualizaci√≥n del histograma

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
quitoImg = cv2.imread("img_lab2/QUITO.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(quitoImg, cmap='gray', vmin=0, vmax=255)
```

```python
hist = cv2.calcHist([quitoImg], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
imagen_equ = cv2.equalizeHist(quitoImg)
plt.imshow(imagen_equ, cmap='gray', vmin=0, vmax=255)
```

```python
hist = cv2.calcHist([imagen_equ], [0], None, [256], [0, 256])
plt.plot(hist)
```

La discontinuidad se deba a que al emparejar los valores lo que est√° ocurriendo es que esta enviando grises a valores lejanos para aumentar el contraste pero esto hace que hayan colores que queden sin pixeles en el proceso.

```python
imagen_equ = cv2.equalizeHist(imagen_equ)
plt.imshow(imagen_equ, cmap='gray', vmin=0, vmax=255)
```

```python
hist = cv2.calcHist([imagen_equ], [0], None, [256], [0, 256])
plt.plot(hist)
```

Lo que ocurre es que la imagen y el histograma se ven iguales, esto se debe a que la ecualizaci√≥n es una operaci√≥n idempotente, es decir que una vez realizada, volver a realizarla no genera cambios nuevos en la im√°gen.

```python
baboonImg = cv2.imread("img_lab2/baboon.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(baboonImg, cmap='gray', vmin=0, vmax=255)
```

```python
hist = cv2.calcHist([baboonImg], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
imagen_equ = cv2.equalizeHist(baboonImg)
plt.imshow(imagen_equ, cmap='gray', vmin=0, vmax=255)
```

```python
hist = cv2.calcHist([imagen_equ], [0], None, [256], [0, 256])
plt.plot(hist)
```

El resultado de esta transformaci√≥n no es tan dr√°stico como el de Quito debido a que la imagen de Baboon ya ten√≠a muchos grises y era una im√°gen buena. Esto hizo que el cambio no fuera tan dr√°stico al hacer la ecualizaci√≥n.
