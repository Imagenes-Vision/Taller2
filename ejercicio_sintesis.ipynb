```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import exposure
```

```python
bougies= cv2.imread('img_lab2/BOUGIES.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(bougies,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([bougies], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
imagen_equ = cv2.equalizeHist(bougies)
plt.imshow(imagen_equ, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
hist = cv2.calcHist([imagen_equ], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
thres_image = cv2.inRange(bougies, 0, 100)
plt.imshow(thres_image,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
negativo= cv2.bitwise_not(thres_image)
plt.imshow(negativo,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
sustract = cv2.subtract(imagen_equ, negativo)
plt.imshow(sustract,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

## Punto 2

```python
muscle = cv2.imread("img_lab2/MUSCLE.png", 0)
plt.imshow(muscle, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
hist= cv2.calcHist([muscle], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
thres_value, thres_image = cv2.threshold(muscle, 85, 255, cv2.THRESH_BINARY)
print(thres_value)
plt.imshow(thres_image,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
negativo = cv2.bitwise_not(thres_image)
plt.imshow(negativo,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
sustract = cv2.add(muscle, negativo)
plt.imshow(sustract,cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```
