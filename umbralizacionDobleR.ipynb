```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
rondelleImg = cv2.imread("img_lab2/RONDELLE.png", cv2.IMREAD_GRAYSCALE)
angioImg = cv2.imread("img_lab2/ANGIO.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(angioImg, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
```

```python
plt.imshow(rondelleImg, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
```

```python
img1 = cv2.inRange(angioImg, 70, 140)
plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
plt.axis("off")
```

```python
# Cambiar el 255 por 128 para los valores entre 70 y 140

array1 = np.where(img1 == 255, 128, 0).astype(np.uint8)
plt.imshow(array1, cmap= "gray", vmin=0, vmax= 255)
plt.axis("off")
```

```python
thres_val, img2 = cv2.threshold(angioImg, 140, 255, cv2.THRESH_BINARY)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
```

```python
result = cv2.bitwise_or(array1, img2)
plt.imshow(result, cmap="gray", vmin=0, vmax=255)
plt.axis("off")
```
