```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

## Calibraci√≥n

```python
quito= cv2.imread('img_lab2/QUITO.png')
plt.imshow(quito, cmap="gray", vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

```python
hist= cv2.calcHist([quito],[0],None,[256],[0,256])
plt.plot(hist)
```

## Expansi√≥n del histograma

```python
rescaled_img = exposure.rescale_intensity(quito, in_range=(np.min(quito),np.max(quito)), out_range=(0, 255)).astype(np.uint8)
plt.imshow(rescaled_img, cmap="gray", vmin=0, vmax=255)
plt.axis("off")
plt.show()
```
